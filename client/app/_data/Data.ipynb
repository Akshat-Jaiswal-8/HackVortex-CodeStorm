```python
import bs4
import requests
```

```python
URL="https://www.geeksforgeeks.org/world-general-knowledge-questions/"
r=requests.get(URL)
print(r.content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(r.content,'html5lib')
print(soup.prettify())
```

```python
import re
import json
myDiv = soup.find_all('p', {'dir': 'ltr'})
all_data = {}
for div in myDiv:
    if 'text-align: justify;' in div.get('style', ''):
        text = div.get_text(strip=True)
        pattern = r'\d+\.\s?([^?]+?)\?-([^\.]+)\.'
        matches = re.findall(pattern, text)
        for match in matches: 
            question = match[0].strip()
            answer = match[1].strip()
            obj = {question: answer}
            y = json.dumps(obj) 
            all_data[question]=answer
            print(all_data)
            with open("Theoretical.json","w") as f:
                json.dump(all_data,f,indent=4)
```
