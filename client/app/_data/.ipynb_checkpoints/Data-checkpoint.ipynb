```python
import bs4
import requests
```

```python
URL="https://www.geeksforgeeks.org/quizzes/trains-boats-and-streams-2-gq/?page=2"
r=requests.get(URL)
print(r.content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(r.content,'html5lib')
print(soup.prettify())
```

```python
import re
import json
myDiv = soup.find_all('p', {'dir': 'ltr'})
all_data = {}
for div in myDiv:
    if 'text-align: justify;' in div.get('style', ''):
        text = div.get_text(strip=True)
        pattern = r'\d+\.\s?([^?]+?)\?-([^\.]+)\.'
        matches = re.findall(pattern, text)
        for match in matches: 
            question = match[0].strip()
            answer = match[1].strip()
            obj = {question: answer}
            y = json.dumps(obj) 
            all_data[question]=answer
            print(all_data)
            with open("Theoretical.json","w") as f:
                json.dump(all_data,f,indent=4)
```

```python
URL2="https://www.geeksforgeeks.org/quizzes/percentages-gq/"
r2=requests.get(URL2)
print(r2.content)
```

```python
soup2=BeautifulSoup(r2.content,'html5lib')
print(soup2.prettify())
```

```python
myDiv2 = soup2.find_all('div', {'class': 'QuizQuestionCard_quizCard__9T_0J'})
allData = []

for div in myDiv2:
    text = div.get_text(separator=" ", strip=True)  # Extract text with a space separator
    if text.startswith("Question"):
        text = text.split(" ", 1)[1]
    allData.append(text)

try:
    with open("Verbal.json", "w") as f:
        json.dump(allData, f, indent=4)
    print("Data successfully written to Verbal.json")
except IOError as e:
    print(f"An error occurred while writing to the file: {e}")
```
